```python
import glob as glob
import pandas as pd
import numpy as np
import pywt
import scipy.signal
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
folders = glob.glob('../../Full data modified/DWT_TextDump/*')
files = []

for fldr in folders:
    files += glob.glob(fldr+'/*.txt')

labels = glob.glob('../../Full data modified/labels/*.txt')
print(len(folders), len(files), len(labels))
```

```python
out_file = open('LabelledTheta.csv', 'w+')
```

```python
out_file.write('FolderName,')
for i in range(1, 15):
    for j in range(38):
        out_file.write(f'Theta_{i}_{j},')
            
out_file.write('Response\n')
out_file.close()
```

```python
out_file = open('./LabelledTheta.csv', 'a')
for k in range(len(folders)):
    fldr_name = folders[k].split('/')[-1]
    
    # Folder name
    out_file.write(f'{fldr_name},')
    
    # Theta values
    for i in range(1, 15):
        data = np.asanyarray(pd.read_csv((folders[k]+f'/Ch{i}.txt'), delimiter='\t', header=None)[3])[:38]
        for j in range(len(data)):
            out_file.write(f'{data[j]},')
    
    # Only for new model idea
    for i in range(1024-532):
        out_file.write(f'0.0,')
        
    # Response
    response = pd.read_csv(f'../../Full data modified/Labels/{fldr_name}.txt', header=None)[0][0]
    out_file.write(f'{response}\n')
    
    print(fldr_name, 'written!')
```
