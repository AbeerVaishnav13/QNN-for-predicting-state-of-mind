```python
import glob as glob
import pandas as pd
import numpy as np
import pywt
import scipy.signal
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
files = glob.glob("Neuromarketing-dataset/Data-EEG/25-users/*.txt")


def get_file_name(file_txt):
    return file_txt.split('/')[3].split('.')[0]

try:
    os.mkdir("./DWT_TextDump")
except:
    pass

len_delta = 38
len_theta = 38
len_alpha = 70
len_beta = 133
len_gamma = 259

idx_d = 4
idx_t = 3
idx_a = 2
idx_b = 1
idx_g = 0
```

```python
print('Making files...')
for file in files:
    data = pd.read_csv(file, delimiter=" ", header=None)
    os.mkdir(f'./DWT_TextDump/{get_file_name(file)}')
    for j in range(14):
        try:
#             print(f'{get_file_name(file)}/Channel{j+1}!', end='\t')

            raw = np.asanyarray(data[j])

            smooth = scipy.signal.savgol_filter(raw,5,2)

            coeffs = pywt.wavedec(smooth,'db4',level=4)

            # store it in a textfile
            target = open(f'./DWT_TextDump/{get_file_name(file)}/Ch{j+1}.txt', 'a')
            for j in range(len(coeffs[4])):
                for i in range(4, -1, -1):
                    if i == 4:
                        target.write(f'{coeffs[i][j]}\t')
                    elif i == 3:
                        if j < len_beta:
                            target.write(f'{coeffs[i][j]}\t')
                        else:
                            target.write('0\t')
                    elif i == 2:
                        if j < len_alpha:
                            target.write(f'{coeffs[i][j]}\t')
                        else:
                            target.write('0\t')
                    elif i == 1:
                        if j < len_theta:
                            target.write(f'{coeffs[i][j]}\t')
                        else:
                            target.write('0\t')
                    elif i == 0:
                        if j < len_delta:
                            target.write(f'{coeffs[i][j]}\n')
                        else:
                            target.write('0\n')
                            
            target.close()
            
        except:
            print(f'{get_file_name(file)}/Channel{j+1} already exists!')
        
print('Done!')
```
